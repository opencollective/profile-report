```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

# Load data from csv
active_users_df = pd.read_pickle('active_users_df.pkl')
all_collectives = pd.read_csv('data/all_collectives.csv', low_memory=False)
# expenses aggregated by collective, including child collectives
# Load data from csv, sql query that generated this csv is queries/expenses_in_period.sql
expenses_df = pd.read_csv('data/expenses_in_period_by_collective.csv', low_memory=False)
# sql query that generated this csv is queries/income_in_period.sql
income_df = pd.read_csv('data/income_in_period_by_collective.csv', low_memory=False)

def camel_to_snake(name):
    s1 = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', name)
    return re.sub('([a-z0-9])([A-Z])', r'\1_\2', s1).lower().replace('_i_d', '_id')

expenses_df.columns = [camel_to_snake(col) for col in expenses_df.columns]
income_df.columns = [camel_to_snake(col) for col in income_df.columns]
all_collectives.columns = [camel_to_snake(col) for col in all_collectives.columns]

# drop expenses and income rows where amount is 0
expenses_df = expenses_df[expenses_df['total_amount'] != 0]
income_df = income_df[income_df['total_amount'] != 0]

# add a new column to expenses_df amountInUSD that is the product of (amountInHostCurrency/100 * exchange) rate for hostCurrency, but if hostCurrency is USD, then amountInUSD is just amountInHostCurrency/100
expenses_df['amount_in_usd'] = expenses_df.apply(lambda row: row['total_amount']/100 * exchange_rates[row['currency']] if row['currency'] != 'USD' else row['total_amount']/100, axis=1)
income_df['amount_in_usd'] = income_df.apply(lambda row: row['total_amount']/100 * exchange_rates[row['currency']] if row['currency'] != 'USD' else row['total_amount']/100, axis=1)
income_df['platform_tip_amount_in_usd'] = income_df.apply(lambda row: row['platform_tip_amount']/100 * exchange_rates[row['currency']] if row['currency'] != 'USD' else row['platform_tip_amount']/100, axis=1)

# add collective slug to expenses_df by merging with all_collectives on collective_id
expenses_df = pd.merge(expenses_df, all_collectives[['id', 'slug']], left_on='collective_id', right_on='id', how='left')
income_df = pd.merge(income_df, all_collectives[['id', 'slug']], left_on='collective_id', right_on='id', how='left')

# rename slug to collective_slug
expenses_df.rename(columns={'slug': 'collective_slug'}, inplace=True)
income_df.rename(columns={'slug': 'collective_slug'}, inplace=True)

# drop id column
expenses_df.drop('id', axis=1, inplace=True)
income_df.drop('id', axis=1, inplace=True)

# add host collective slug to expenses_df by merging with all_collectives on host_collective_id
expenses_df = pd.merge(expenses_df, all_collectives[['id', 'host_fee_percent', 'slug']], left_on='host_collective_id', right_on='id', how='left')
income_df = pd.merge(income_df, all_collectives[['id', 'host_fee_percent', 'slug']], left_on='host_collective_id', right_on='id', how='left')

# rename slug to host_collective_slug
expenses_df.rename(columns={'slug': 'host_collective_slug'}, inplace=True)
income_df.rename(columns={'slug': 'host_collective_slug'}, inplace=True)

# drop id column
expenses_df.drop('id', axis=1, inplace=True)
income_df.drop('id', axis=1, inplace=True)

# reorder columns in expenses_df like this: collective_id, currency, transaction_count, total_amount, amount_in_usd
expenses_df = expenses_df[['collective_id', 'collective_slug', 'host_collective_id', 'host_collective_slug', 'host_fee_percent', 'transaction_count', 'amount_in_usd']]
income_df = income_df[['collective_id', 'collective_slug', 'host_collective_id', 'host_collective_slug', 'payment_method_type', 'host_fee_percent', 'transaction_count', 'amount_in_usd', 'platform_tip_amount_in_usd']]

# merge rows in expenses_df where 'collective_id', 'collective_slug', 'host_collective_id', 'host_fee_percent', 'host_collective_slug' are the same, and sum 'transaction_count' and 'amount_in_usd'
expenses_df = expenses_df.groupby(['collective_id', 'collective_slug', 'host_collective_id', 'host_collective_slug', 'host_fee_percent', ]).agg({'transaction_count': 'sum', 'amount_in_usd': 'sum'}).reset_index()
income_df = income_df.groupby(['collective_id', 'collective_slug', 'host_collective_id', 'host_collective_slug', 'payment_method_type', 'host_fee_percent']).agg({'transaction_count': 'sum', 'amount_in_usd': 'sum' , 'platform_tip_amount_in_usd': 'sum'}).reset_index()

# format amount_in_usd as int
expenses_df['amount_in_usd'] = abs(expenses_df['amount_in_usd'].astype(int))
income_df['amount_in_usd'] = abs(income_df['amount_in_usd'].astype(int))
income_df['platform_tip_amount_in_usd'] = abs(income_df['platform_tip_amount_in_usd'].astype(int))

# if transaction_count for expenses is less than 5 and collective_id is same as host_collective_id, then drop that row and save the collective_ids that were dropped in a list
# Getting the collective_id values of rows that meet the drop condition
dropped_collective_ids = expenses_df.loc[(expenses_df['transaction_count'] < 5) & (expenses_df['collective_id'] == expenses_df['host_collective_id']), 'collective_id']

# Apply the filter to retain the rows in expenses_df that don't meet the drop condition
expenses_df = expenses_df[~((expenses_df['transaction_count'] < 5) & (expenses_df['collective_id'] == expenses_df['host_collective_id']))]

# Drop same rows from income_df
income_df = income_df[~income_df['collective_id'].isin(dropped_collective_ids)]

# drop collectives with slug ending with <string>-<number>.<number> as this is a known spammer pattern for some reason
expenses_df = expenses_df[~expenses_df['collective_slug'].str.contains(r'-\d+\.\d+$')]
income_df = income_df[~income_df['collective_slug'].str.contains(r'-\d+\.\d+$')]

# drop internal opencollective collectives where host_collective_slug is 'opencollective'
expenses_df = expenses_df[~(expenses_df['host_collective_slug'] == 'opencollective')]
income_df = income_df[~(income_df['host_collective_slug'] == 'opencollective')]

# Drop rows from income_df if the sum of all transaction_count for rows with that collective_id in income_df is less than 5.

# 1. Group by 'collective_id' and get sum of 'transaction_count' for each 'collective_id'
transaction_sums = income_df.groupby('collective_id')['transaction_count'].sum()

# 2. Filter out 'collective_id's where the sum of 'transaction_count' is less than 5
low_transaction_ids = transaction_sums[transaction_sums < 5].index

# 3. Group by 'collective_id' and check if all 'payment_method_type' for each 'collective_id' is 'host'
all_host = income_df.groupby('collective_id')['payment_method_type'].apply(lambda x: all((val == 'host' or val == 'blank') for val in x))

# 4. Filter out 'collective_id's where all 'payment_method_type' is 'host'
host_payment_ids = all_host[all_host].index

# 5. Combine 'collective_id's based on the conditions and their absence in 'expenses_df'
dropped_collective_ids_from_low_transactions = set(low_transaction_ids) - set(expenses_df['collective_id'])
dropped_collective_ids_from_host_payments = set(host_payment_ids) - set(expenses_df['collective_id'])
dropped_collective_ids = dropped_collective_ids_from_low_transactions.union(dropped_collective_ids_from_host_payments)

# 6. Drop rows from 'income_df' that have 'collective_id's in 'dropped_collective_ids'
income_df = income_df[~income_df['collective_id'].isin(dropped_collective_ids)]
```
